```python
df = pd.read_csv('data_Preprocessed_Onehot.csv')
```

```python
df.head()
```

```python
df.dtypes
```

# OPTICS Clustering

```python
clusterer = OPTICS(min_samples=0.05, n_jobs=-1)

cluster_labels = clusterer.fit_predict(df)
```

# Evaluation

```python
# Silhouette score
score = silhouette_score(df, cluster_labels)
print('Silhouette score of OPTICS clustering: ', score)
```

```python
# Save to csv
cluster = pd.DataFrame(data=cluster_labels, columns=['cluster'])
cluster.to_csv('clusterLabel_OPTICS.csv', index=False)
```
